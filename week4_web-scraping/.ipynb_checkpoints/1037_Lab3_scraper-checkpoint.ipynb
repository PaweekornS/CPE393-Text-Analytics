# Web Scraper

````python
from bs4 import BeautifulSoup
import requests
````

Require input is the URL.

````python
url = 'https://www.britannica.com/topic/list-of-state-capitals-in-the-United-States-2119210'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
print(soup)
````

Find all tables in the page. In this website, though, there's only one so it's simple.

````python
soup.find_all('table')
````

Let's first get just the title of the table.

*Note: `<th>` tag defines a header cell in an HTML table*

````python
titles = soup.find_all('th')
titles
````

Since we do not need the tags, let's clean up the data.

````python
titles_list = [title.text for title in titles]
titles_list
````

If the output still contains newline and other symbols that are not needed, you can further clean the data using, for example, ```.strip()```

Next, create a dataframe

````python
import pandas as pd

df = pd.DataFrame(columns = titles_list)
df
````

Let's scrape the remaining data and fill this table!

````python
rows = soup.find_all('tr')
len(rows)
````

The data of our interest are within the scope of **td** tags.

*Note: `<td>` tag defines a standard data cell in an HTML table.*

````python
# -- Long version --
# row_data = []
# for row in rows:
#   row_data.append(row.find_all('td'))

# -- Short version --
row_data = [row.find_all('td') for row in rows]
row_data
````

The first row collected has no value, thus an empty list.

````python
row_data[0]
````

Turns out the first row is actually here

````python
row_data[1]
````

However, we only want the text portion.

````python
print(row_data[1][0].text)
print(row_data[1][1].text)
print(row_data[1][2].text)
print(row_data[1][3].text)
````

Therefore, more cleaning is necessary.

Add the remaining rows to the dataframe.

But does this code work?

````python
row_data
````

````python
for sentence in row_data[1:]:
    info=[]
    for elem in sentence:
        info.append(elem.text)
    
    df.loc[len(df)] = info
````

````python
df.drop(0, inplace=True)
df
````

````python
url = 'https://en.wikipedia.org/wiki/List_of_Bandai_Namco_video_games'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
soup
````

````python
table = soup.find_all('th')

all_title = [title.text.strip('\n') for title in table]
all_title
````

````python
# I use slicing because this wikipedia has 2 tables which is written by same html tags format: 'Video games' and 'Mobile games'
# I will focus on the 'Video games' table.
titles = all_title[:5]
titles
````

````python
rows = soup.find_all('tr')
rows_data = [row.find_all('td') for row in rows]

game_info = [result for result in rows_data if len(result) == 5]
game_info
````

````python
bandai = pd.DataFrame(columns=titles)

ind = 0
for record in game_info:
    temp = []
    for elem in record:
        temp.append(elem.text.strip('\n'))
        
    bandai.loc[ind] = temp
    ind += 1
    
    
bandai
````

### Clean & Preprocess

````python
bandai.drop(columns='Ref(s)', inplace=True)
bandai.head()
````

````python
# word tokenization
test_df = bandai.copy()

ind = 0
for platform in test_df['Platforms']:
    test_df.loc[ind, 'Platforms'] = platform.split(', ')
    ind += 1
    
test_df
````

I think this dataset needs to preprocess by text tokenization only because it didn't contain the sentence or verb in any columns. I will do the text visualization for extra because I haven't do any much at this point. :)

<b>Text Visualization</b>

````python
from itertools import chain

platforms = list(test_df['Platforms'].values)

platforms_list = []
for list_elem in platforms:
    platforms_list = list(chain(platforms_list, list_elem))
    
platforms_list[:5]
````

````python
from nltk.probability import FreqDist

word_freq = FreqDist(platforms_list)

# Convert word frequencies to a DataFrame for seaborn
data = {'Word': list(word_freq.keys()), 'Frequency': list(word_freq.values())}
df_word_freq = pd.DataFrame(data)

# Sort DataFrame by frequency in descending order
df_word_freq = df_word_freq.sort_values(by='Frequency', ascending=False)

df_word_freq.head(10)
````

````python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

text = ' '.join(platforms_list)
# Generate word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Display the generated word cloud using matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
````

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate word frequencies
word_freq = FreqDist(platforms_list)

# Convert word frequencies to a DataFrame for seaborn
data = {'Word': list(word_freq.keys()), 'Frequency': list(word_freq.values())}
df_word_freq = pd.DataFrame(data)

# Sort DataFrame by frequency in descending order
df_word_freq = df_word_freq.sort_values(by='Frequency', ascending=False)

# Plot a bar chart using seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x='Word', y='Frequency', data=df_word_freq.head(20), palette='viridis', hue='Word')
plt.title('Top 20 Most Frequent Words')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
````
